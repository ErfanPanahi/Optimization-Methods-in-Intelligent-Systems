```python
#@title Functions
def opt_J_pack(X,Y,etha,ITR,num):
  [N,L] = X.shape
  b = 0
  w = np.zeros(L)
  J_values = [J(b,w,X,Y)]
  for itr in range (ITR):
    rnd = random.sample(range(N),100)
    b = b - etha * grad_b_J(b,w,X[rnd,:],Y[rnd])
    w = w - etha * grad_w_J(b,w,X[rnd,:],Y[rnd])
    J_values.append(J(b,w,X,Y))
    if J(b,w,X,Y) < 1e-2:
      break
  return b,w,J_values  

def ACC(b,w,X,Y_test):
  y_hat = []
  [N,L] = X.shape
  acc = 0
  for i in range(N):
    y_hat.append(np.sign(b + np.dot(w,X[i,:])))
    if y_hat[i] == Y_test[i]:
      acc += 1/N
  return acc

def opt_J(X,Y,etha,ITR):
  [N,L] = X.shape
  b = 0
  w = np.zeros(L)
  J_values = [J(b,w,X,Y)]
  for itr in range (ITR):
    b = b - etha * grad_b_J(b,w,X,Y)
    w = w - etha * grad_w_J(b,w,X,Y)
    J_values.append(J(b,w,X,Y))
    if J(b,w,X,Y) < 1e-2:
      break
  return b,w,J_values
def mu(b,w,xi,yi):
    mu_i = 1 / (1 + np.exp(-yi * (b + np.dot(w,xi))))
    return mu_i
def grad_b_mu(b,w,xi,yi):
    grad_mu  = yi * np.exp(-yi * (b + np.dot(w,xi))) / (1 + np.exp(-yi * (b + np.dot(w,xi))))**2
    return grad_mu
def grad_w_mu(b,w,xi,yi):
    grad_mu  = xi * yi * np.exp(-yi * (b + np.dot(w,xi))) / (1 + np.exp(-yi * (b + np.dot(w,xi))))**2
    return grad_mu
def J(b,w,X,Y):
    [N,L] = X.shape
    Jwb = 0
    for i in range(N):
        Jwb += np.log((1 + np.exp(-Y[i] * (b + np.dot(w,X[i,:]))))) / N
    return Jwb
def grad_b_J(b,w,X,Y):
    [N,L] = X.shape
    grad_J = 0
    for i in range(N):
        # grad_J += -(grad_b_mu(b,w,X[i,:],Y[i]) / mu(b,w,X[i,:],Y[i])) / N
        grad_J += - Y[i] * (1 - mu(b,w,X[i,:],Y[i])) / N
    return grad_J
def grad_w_J(b,w,X,Y):
    [N,L] = X.shape
    grad_J = 0
    for i in range(N):
        # grad_J += -(grad_w_mu(b,w,X[i,:],Y[i]) / mu(b,w,X[i,:],Y[i])) / N
        grad_J += - X[i,:] * Y[i] * (1 - mu(b,w,X[i,:],Y[i])) / N
    return grad_J
```
